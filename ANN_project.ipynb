```python
#Encoding the categorical values
label=LabelEncoder()
data['Gender']=label.fit_transform(data['Gender'])
```

```python
#I need to convert geography using one hot encoding
from sklearn.preprocessing import OneHotEncoder
one_label=OneHotEncoder()
one_encoded=one_label.fit_transform(data[['Geography']])
```

```python
one_encoded.toarray()
```

```python
columns = one_label.get_feature_names_out(['Geography'])
columns
```

```python
geo_encoded=pd.DataFrame(one_encoded.toarray(),columns = one_label.get_feature_names_out(['Geography'])
)
```

```python
geo_encoded
```

```python
#Combine all the one hot encoded columns
data=pd.concat([data.drop('Geography',axis=1),geo_encoded],axis=1)
```

```python
data
```

```python
#Divide the data set into dependent and independent
x=data.drop('Exited',axis=1)
y=data["Exited"]
```

```python
y
```

```python
x
```

```python
# i want to split my data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Scaling the features
salar=StandardScaler()
x_test=salar.fit_transform(x_test)
x_train=salar.transform(x_train)
```

APPLYING ANN IMPLEMENTATION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import datetime
```

```python
!pip install tensorboard
```

```python
#Building an ANN Model
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
#compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir='logs/fit'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Set up early Stopping
early_stopping_callback=EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
#training the neural network model
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                  callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
model.save("model.h5")
```
